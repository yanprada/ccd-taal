```python
import pandas as pd
```

```python
original_filename = 'anamnese_sentencas.xlsx'
corrected_filename = 'anamnese_sentencas_corrigidas.xlsx'

df_original = pd.read_excel(original_filename, sheet_name=None)
df_corrected = pd.read_excel(corrected_filename, sheet_name=None)
```

```python
inconsistencies = []
for sheet_name in df_original.keys():
    df_orig = df_original[sheet_name]
    df_corr = df_corrected[sheet_name]

    for row in range(len(df_orig)):
        id_orig = df_orig.at[row, 'ID']
        id_corr = df_corr.at[row, 'ID']
        if isinstance(id_orig, str) and id_orig != id_corr:
            inconsistencies.append((sheet_name, row, id_orig, id_corr))
```

```python
print(f"Found {len(inconsistencies)} inconsistencies:")
for inconsistency in inconsistencies:
    sheet, row, id_orig, id_corr = inconsistency
    print(f"Inconsistency in sheet '{sheet}', row {row}: original ID '{id_orig}' vs corrected ID '{id_corr}'")
```
